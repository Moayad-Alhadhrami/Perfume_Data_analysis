```python
Sales_tran['Sales']=Sales_tran['Price']*Sales_tran['Quantity']
```

```python
Sales_tran['Profit']=(((Sales_tran['Price']*Sales_tran['Quantity'])*(Sales_tran['Discount']-100))-(Sales_tran['Cost Price']))
```

```python
pd.options.display.float_format = '{:.0f}'.format
```

```python
Product_by_gender=Sales_tran.groupby('Gender')['Sales','Profit'].sum().sort_values(by='Sales',ascending=False)
Product_by_gender
```

```python
Sales_by_year=Sales_tran.groupby('Year')[['Sales']].sum()
Sales_by_year
```

```python
Sales_by_Year_Qu=Sales_tran.groupby(['Year','Quarter'])[['Sales']].sum()
Sales_by_Year_Qu
```
